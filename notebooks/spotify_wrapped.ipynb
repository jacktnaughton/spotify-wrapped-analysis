```python
import pandas as pd
import json
import os

folder1 = 'data'
all_data_1 = []

for i in range(2):
    filename = f'StreamingHistory_music_{i}.json'
    filepath = os.path.join(folder1, filename)
    
    with open(filepath, 'r') as f:
        data = json.load(f)
        df = pd.json_normalize(data)
        all_data_1.append(df)

combined_df_1 = pd.concat(all_data_1, ignore_index=True)
```

```python
csv_path = "data/year_in_streaming.csv"
combined_df_1.to_csv(csv_path, index=False)
```
